```python
def encode_albums(table):
    """
    Converts a BeautifulSoup object into individual album lists
    with text changed in order to properly use as URL strings.
    """    

    albums_temp = []

    # A loop to pull out titles for each album
    for idx, element in enumerate(table):
        title = element.get_text()
        title = title.split('\n')[0]
        albums_temp.append(title)

    # Encoding the text into UTF-8 by hand becuase I could not figure
    # out .encode/.decode
    albums = []
    for album in albums_temp:
        album = (album
                .replace("'", '%27')
                .replace(",", '%2C')
                .replace("?", '%3F')
                .replace("[", '%5B')
                .replace("]", '%5D')
                .replace("(", '%28')
                .replace(")", '%29')
                .replace("*", '%2A')
                .replace("&", '%26')
                .replace("|", '%7')
                .replace(" ", '-')
                .replace("!", '%21')
                )
        albums.append(album)

    return albums 
```

```python
def find_solo_albums(artist):
    """
    This function searches Wikipedia for an artist's discography page. 
    Unfortunately, Wikipedia is not %100 consistent with their naming.
    Therefore, 3 functions are required to thoroughly search, in some
    case, this is still not enough.

    Args:
        artist (str): The name of an artist to search for. Can be written
        with spaces, or as snake case, can be all caps, lower, or title.

    Returns:
        albums (list): A list of all the albums written and released by 
        the artist. This method ensures that the albums listed are those
        which the artist is the main artist, and not simply a guest or
        featured artist.
    """
    # Create a path to save each artist's albums/lyrics
    path = f"{artist.replace(' ','_')}"
    
    if not os.path.exists(path):
        os.makedirs(path)

    # Change format of artist name in order to get to the right
    # Wikipedia format   
    artist = artist.title().replace(' ','_')
    
    URL = f'https://en.wikipedia.org/wiki/{artist}_discography'
    
    # Web-scrapig
    main_page = requests.get(url=URL)
    wiki = main_page.text
    filename = f'{artist}_wiki.txt'
    
    with open(os.path.join(path, filename), mode="w") as file:
        file.write(wiki)
    
    soup = BeautifulSoup(wiki, 'html.parser')
    
    # Searching the Wikipedia pages for a specific table which exists 
    # in most discography pages (however, not all...)
    if not soup.find('table',{'class':'wikitable plainrowheaders'}):    
        print('Artist Discography Page not Found:\n Try find_solo_albums_alt')
    
    else:
        table = (soup
                   .find('table',{'class':'wikitable plainrowheaders'})
                   .find('tbody')
                   .find_all('th',{'scope':'row'})
                  )
    
        return encode_albums(table)
```

```python
def find_solo_albums_alt(artist):
    """
    The only difference between these functions is in which table 
    BeautifulSoup searches.

    Args:
        artist (str): The name of an artist to search for. Can be written
        with spaces, or as snake case, can be all caps, lower, or title.

    Returns:
        albums (list): A list of all the albums written and released by 
        the artist.
    """
    # Everything here is the same except for what BeautifulSoup
    # searches for
    path = f"{artist.replace(' ','_')}"
    
    if not os.path.exists(path):
        os.makedirs(path)
        
    artist = artist.title().replace(' ','_')
    
    URL = f'https://en.wikipedia.org/wiki/{artist}_discography'
    
    main_page = requests.get(url=URL)
    wiki = main_page.text
    filename = f'{artist}_wiki.txt'
    
    with open(os.path.join(path, filename), mode="w") as file:
        file.write(wiki)
    
    soup = BeautifulSoup(wiki, 'html.parser')
    
    # This is an alternate table that some discography pages use
    # However, there are more than two. Trying to match all of them
    # was impossible for this week and perhaps Selenium would be a 
    # better solution.
    if not soup.find('table',{'class':'wikitable'}):    
        print('Artist Discography Page Still not Found:\n Try find_solo_albums_alt2')
    
    else:
        table = (soup
                .find('table',{'class':'wikitable'})
                .find_all('i')
                )

        return encode_albums(table)
```

```python
def find_solo_albums_alt2(artist):
    """
    The only difference between these functions is in which table 
    BeautifulSoup searches.

    Args:
        artist (str): The name of an artist to search for. Can be written
        with spaces, or as snake case, can be all caps, lower, or title.

    Returns:
        albums (list): A list of all the albums written and released by 
        the artist.
    """
    path = f"{artist.replace(' ','_')}"
    
    if not os.path.exists(path):
        os.makedirs(path)
 
    artist = artist.title().replace(' ','_')
    
    # In this function, this is the difference. The tail of this string
    # is different than the others.
    URL = f'https://en.wikipedia.org/wiki/{artist}_albums_discography'
    
    main_page = requests.get(url=URL)
    wiki = main_page.text
    filename = f'{artist}_wiki.txt'
    
    with open(os.path.join(path, filename), mode="w") as file:
        file.write(wiki)
    
    soup = BeautifulSoup(wiki, 'html.parser')
    
    if not soup.find('table',{'class':'wikitable plainrowheaders'}):    
        print('Artist Discography Page Really not Found:\n Dang.')
    
    else:
        table = (soup
                   .find('table',{'class':'wikitable plainrowheaders'})
                   .find('tbody')
                   .find_all('th',{'scope':'row'})
                  )

        return encode_albums(table)
```

# Now, let's take those chosen albums and find all the lyrics from them

* These functions went through many iterations before this final form, including a form where every function that follows was a part of one larger function.
* It was large, in charge and beautiful...but hard to make sense of.

#### I wanted this function to take in as little as possible and still function, however, I also wanted it to run rather quickly, so I decided that it should be the users responsibility to find the URL for their artists' lyrics.com page, as well as define a list of predetermined albums to include.

```python
def make_artist_path(url):
    """ 
    This function uses a given URL and returns the artist's name
    after cleaning and the path that will later be used for labeling 
    and storing an artist's song collection.

    Args:
        url (_str_): _An artist's home page on lyrics.com_
    
    Returns:
        _str_: The artist's name, in title format with -'s 
        replaced with _'s.
        _str_: The artist's name, in lower format with
        -'s replaced with _'s.
    """

    # Create the artists name using the URL
    artist = url.split('/')[-2]
    
    # Establish a path to save albums
    artist_path = f"{artist.lower().replace('-','_')}"
    
    if not os.path.exists(artist_path):
        os.makedirs(artist_path)
    
    # Replace any - with _
    artist = artist.lower().replace('-', '_')

    return artist, artist_path
```

```python
def make_album_list(url, artist, artist_path):
    """
    Creates a txt file that contains all of the HTML information
    for an artist's lyrics.com homepage. Also creates a list of 
    every album that can be found there.

    Args:
        url (_str_): _Homepage of an artist at lyrics.com_
        artist (_string_): _Name of an artist defined by 
        make_artist_path_
        artist_path (_string_): _Folder location, defined by 
        make_artist_path_

    Returns:
        -.txt file_: A file saved in the newly created artist_path
        directory. Contains HTML output of artist's lyrics.com
        homepage.
        _list_: _A listing of every album found on an artist's 
        lyrics.com homepage._
    """

    main_page = requests.get(url)
    home_page = main_page.text
    filename = f'{artist}_main.txt'

    # Create the txt files to later search through
    with open(os.path.join(artist_path, filename), mode="w") as file:
        file.write(home_page)

    with open(os.path.join(artist_path, filename), mode='r') as file:
        first_page = file.read()
        
    # Find all the album titles using RegEx
    album_list = re.findall(r'/album/[^"]+', first_page)

    return album_list
```

```python
def make_album_sites_names(albums, album_list):
    """
    Takes in a user defined list of albums to include, and the list
    of albums defined by make_album_list and creates a list of 
    websites to later scrape.

    Args:
        albums (_list of strings_): _A list of albums chosen by the user
        to include in an artist's catalogue._
        album_list (_list of strings_): _A list of albums created by
        make_album_list, which includes every album listed at lyrics.com._

    Returns:
        _list of strings_: _A list of URLs that will be used in scraping._
        _list of strings_: _A list of album names that will create directories._
    """
    # Empty list to store all found album websites
    album_sites = []
    album_names = []
        # Goes through each album found in the first_page file, checks for duplicates,
        # appends a link to make the album a website for later use, and adds the 
        # cleaned name to a list to check against duplicates.
    for album in album_list:
        album_test = album.split('/')[-1]
        if album_test in albums:
            album_sites.append('https://www.lyrics.com'+album)
            album_names.append(album_test)   
        else:
            pass
    
    return album_sites, album_names
```

```python
def create_album_path_page(artist_path, site):
    """
    Creates directories for each album of an artist and returns
    both a file to decipher with BeautifulSoup and a path to that
    file.

    Args:
        artist_path (_str_): _A path to a folder, defined in make_artist_path_
        site (_type_): _The website for a specific album. This is defined
        as part of a for loop in find_lyrics. Each site is from album_sites.

    Returns:
        _txt file_: _An HTML file containing site information on one album f
        or the given artist_
        _str_: _A path for each album inside the larger folder of the artist.
    """

    # Defining path to save albums
    album_path = f"{artist_path}/{site.split('/')[-1]}"

    if not os.path.exists(album_path):
        os.makedirs(album_path)
        
    # Scrape each album website and get the full text
    album_request = requests.get(site)
    album_site = album_request.text
    album_txt = f"{site.split('/')[-1]}.txt"
    
    # Create new files to store Albums
    with open(os.path.join(album_path, album_txt), mode="w") as file:
        file.write(album_site)
    
    # Opening each newly created Album file
    with open(os.path.join(album_path, album_txt), mode='r') as file:
        album_page = file.read()
    
    return album_page, album_path
```

```python
def fix_utf8(song_name):
    """
    Takes in a string containing the name of a song and cleans
    it. Roughly equivalent to .encode/.decode.

    Args:
        song_name (_str_): _A song title in UTF-8 format_

    Returns:
        _str_: _The same title with replacements for characters_
    """
    song_name = (song_name
                .replace("+"," ")
                .replace('%27', "'")
                .replace('%2C', ",")
                .replace('%3F', "?")
                .replace('%5B', "[")
                .replace('%5D', "]")
                .replace('%28', "(")
                .replace('%29', ")")
                .replace('%2A', "*")
                .replace('%26', "&")
                .replace('%7', "|")
                .replace('%21', "!")
                )

    return song_name
```

```python
def fill_songs(song_dictionary, song_sites, album_path) -> None: 
    """
    Using a list of song URLs, created in the find_lyrics function,
    this function scrapes each page for lyrics and updates a predefined
    dictionary with key/value pairs such that each song is a key and the 
    lyrics a value. Duplicates are accounted for unless the song title 
    is less than 4 characters long.

    Args:
        song_dictionary (_dict_): _An empty dictionary defined in find_lyrics_
        song_sites (_list of str_): _A list of song sites defined in 
         find_lyrics. _
        album_path (_str_): _The path to each specific album, defined in 
        create_album_path_page_

    Returns:
        None: Fills a predefined empty dictionary.
    """

    # Creating an empty list to check for duplicates
    track_list = []
    
    # Creating an empty list to add to a dictionary
    song_title = []
    
    # Loop through every song in the song list
    for song in song_sites:
        
        # Specify where to save the songs
        song_path = f"{album_path}"

        # Beginning to scrape through each song URL 
        song_request = requests.get(url=song)
        song_html = song_request.text
        song_txt = f"{song.split('/')[-1]}.txt"
        song_name = f"{song.split('/')[-1]}"

        song_name = fix_utf8(song_name)

        # Use Beautiful Soup for easier parsing
        soup = BeautifulSoup(song_html, 'html.parser')
        
        # If soup does not find this in the song HTML, the lyric text is empty
        if not soup.find(name='pre'):
            pass
        else:
            lyrics = soup.find(name='pre').get_text()

        # If the first characters of a song name are already in the track list,
        # skip the song.
        if song_txt[0:5] in track_list:
            pass
        else:
            track_list.append(song_txt[0:5])
            song_title.append(song_name)
            song_dictionary[song_name] = lyrics
            
            with open(os.path.join(song_path, song_txt), mode="w") as file:
                file.write(lyrics)
                
        time.sleep(1)
```

```python
def create_df(song_dictionary, artist):
    """
    Takes in a dictionary and artist name and returns a DataFrame with 
    columns: 'artist','song','lyrics'. Used for building a text analyzer.

    Args:
        song_dictionary (_dict_): _A dictionary containing songs and lyrics 
        as key/value pairs._
        artist (_str_): _The name of an artist, defined in make_artist_path_

    Returns:
        _DataFrame_: _A DataFrame with columns: 'artist','song','lyrics'._
    """

    # Creating a DataFrame that conatins the artist, songs, and cleaned lyrics
    song_df = (pd
               .DataFrame(dict([(k,pd.Series(v)) for k,v in song_dictionary.items()]))
               .T
               .reset_index()
              )
    song_df.insert(0, 'artist', artist)
    song_df.rename(columns={'artist': "artist", 'index': "song", 0: "lyrics"}, 
                   inplace=True)
    song_df['lyrics'] = (song_df['lyrics']
                         .replace('(\\r\\n)',' ', regex=True)
                         .replace('[^a-zA-z ]',' ', regex=True)
                         .replace('(\s\s+)',' ', regex=True)
                         .str.lower()
                        )
     
    return song_df
```

```python
def find_lyrics(url, albums):
    """
    Takes in a user defined URL and list of selected albums, then uses 
    those to scrape lyrics.com for specific lyrics for the given artist.

    Args:
        url (_str_): _The lyrics.com homepage for an artist. Be sure to 
        exactly copy paste or scraping will not be successful._
        albums (_list of strings_): _A list of specific albums that the user
        wishes to use. Must be in UTF-8 format. Use find_solo_albums for help
        in creating this list._

    Returns:
        _DataFrame_: _A DataFrame with columns: 'artist','song','lyrics'._
    """

    artist, artist_path = make_artist_path(url)
    album_list = make_album_list(url, artist, artist_path)
    album_sites, album_names = make_album_sites_names(albums, album_list)

    # Creating an empty song dictionary for later DataFrame use
    song_dictionary = {}
    
    # Running a loop through the newly created list of Album URLs
    for site in album_sites:
        
        album_page, album_path = create_album_path_page(artist_path, site)
        
        # Pulling out the song titles from the larger HTML text
        song_ends = re.findall(r'/lyric/[^"]+', album_page)
        song_sites = []
        
        # Making the song ends into full URLs
        for song in song_ends:
            song_sites.append('https://www.lyrics.com'+song)
        
        time.sleep(1)

        fill_songs(song_dictionary, song_sites, album_path)
    
    song_df = create_df(song_dictionary, artist)    
    
    return song_df
```

```python
def make_pickle(df, name):
    """
    Creates a pickle file from a DataFrame

    Args:
        df (_DataFrame_): _A DataFrame to create the pickle file from_
        name (_str_): _A user chosen name for the file_
    """
    file = open(name, 'wb')
    pickle.dump(df, file)
    file.close()
```
