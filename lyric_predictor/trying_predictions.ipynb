```python
import pickle

import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.linear_model import LogisticRegression
```

```python
# Opening the five_artists pickle file that has lyrics and artists
file = open('five_artists', 'rb')

full_list = pickle.load(file)

file.close()
```

```python
# I chose 5 artists with very distinct styles so my predictions should be
# fairly accurate.

full_list['artist'].unique()
```

```python
CORPUS = full_list['lyrics'].to_list()
CORPUS
```

```python
LABELS = full_list['artist'].to_list()
LABELS
```

* I will clean the artist names later on when printing the prediction.

```python
df_lyrics = pd.DataFrame(CORPUS, index=LABELS, columns=['lyrics'])
df_lyrics
```

```python
X = df_lyrics['lyrics']
y = df_lyrics.index
```

* In order to pick out more specific trends in the lyrics, I must first define stop words.

```python
cv = CountVectorizer(stop_words='english',ngram_range=(1,1))
```

```python
X_tran = cv.fit_transform(X)
```

```python
X_tran.shape
```

```python
print(X_tran)
```

```python
X_tran.todense()
```

```python
cv.get_feature_names_out()
```

```python
pd.DataFrame(X_tran.todense(),columns=cv.get_feature_names_out(),index=y)
```

* Use Tfidf Transformation to find the specific weights of importance for each word in each lyrical text.

```python
tf = TfidfTransformer()
```

```python
tf_X_tran = tf.fit_transform(X_tran)
```

```python
tf_X_tran.todense()
```

```python
pd.DataFrame(tf_X_tran.todense(), columns=cv.get_feature_names_out(),index=y)
```

```python
lreg = LogisticRegression()
```

```python
lreg.fit(tf_X_tran,y)
```

```python
def predict_it(lyrics):
    """
    Predicts which artist can be attributed to a user given lyric.

    Args:
        lyrics (_str_): _A string of proposed lyrics for the model to predict upon._
    """
    lyrics_to_predict = [lyrics]

    lyrics_to_predict_tran = cv.transform(lyrics_to_predict)

    X_pred_trans = tf.transform(lyrics_to_predict_tran)

    # Some of my artist names are encoded in UTF-8
    # However, .encode/.decode were not working well so I did it by hand
    prediction = (lreg
                    .predict(X_pred_trans)[0]
                    .replace('%27', "'")
                    .replace('%2c', ',')
                    .replace('%3f', '?')
                    .replace('%5b', '[')
                    .replace('%5d', ']')
                    .replace('%28', "(")
                    .replace('%29', ")")
                    .replace('%2a', "*")
                    .replace('%26', "&")
                    .replace('%7', "|")
                    .replace('-', " ")
                    .replace('%21', "!")
                    .replace('_', ' ')
    )
                    
    print(f'My guess is: {prediction}')
```

```python
predict_it('do you remember the night of september')
```
